```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, cohen_kappa_score
```

```python
# load excel results predicted using GPT models from last step
df=pd.read_excel('1 gpt4o_results.xlsx')
df
```

```python
# Define the dimension
dimensions = ["attractiveness"]

# Dictionary to store results
evaluation_results = {}

# Set plot style
sns.set(style="whitegrid")

# Compute evaluation metrics for each dimension
for dim in dimensions:
    y_true = df[dim]
    y_pred = df[f'gpt_attractiveness__4o']# here use the predicted perception column name

    accuracy = accuracy_score(y_true, y_pred)
    conf_matrix = confusion_matrix(y_true, y_pred)
    class_report = classification_report(y_true, y_pred, digits=4)
    kappa_score = cohen_kappa_score(y_true, y_pred)

    # Store results
    evaluation_results[dim] = {
        "Accuracy": accuracy,
        "Cohen Kappa Score": kappa_score,
        "Classification Report": class_report
    }

    # Plot Confusion Matrix
    plt.figure(figsize=(6,5))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
                xticklabels=[0,1,2], yticklabels=[0,1,2])
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.title(f"Confusion Matrix - {dim.capitalize()}")
    plt.savefig('label.png',dpi=300)
    plt.show()

```

```python
# Print evaluation results
for dim, metrics in evaluation_results.items():
    print(f"\nEvaluation Metrics for {dim.capitalize()}:")
    print(f"Accuracy: {metrics['Accuracy']:.4f}")
    print(f"Cohen Kappa Score: {metrics['Cohen Kappa Score']:.4f}")
    print(f"Classification Report:\n{metrics['Classification Report']}")
```
